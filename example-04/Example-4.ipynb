```python
import pyxtal
print(pyxtal.__version__)
```

```python
from pyxtal.db import database
db = database('../dataset/mech.db')

code1, code2, code3, code4 = 'BIYRIM01', 'DAHMUX', 'DAHLOQ', 'YEWYAD'

xtal1 = db.get_pyxtal(code1)
sph1 = xtal1.get_spherical_images()

xtal2 = db.get_pyxtal(code2)
sph2 = xtal2.get_spherical_images()

xtal3 = db.get_pyxtal(code3)
sph3 = xtal3.get_spherical_images()

xtal4 = db.get_pyxtal(code4)
sph4 = xtal4.get_spherical_images()

sph1.plot_sph_images() #molecule=True)
sph2.plot_sph_images() #molecule=True)
sph3.plot_sph_images() #molecule=True)
sph4.plot_sph_images() #molecule=True)
```

```python
S12 = sph1.get_similarity(sph2, M=10, cutoff=0.95).mean(); print(S12)
S13 = sph1.get_similarity(sph3, M=10, cutoff=0.95).mean(); print(S13)
S14 = sph1.get_similarity(sph4, M=10, cutoff=0.95).mean(); print(S14)
```
